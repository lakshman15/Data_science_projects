```
import pandas as pd
import numpy as np
ed  = pd.read_csv("https://raw.githubusercontent.com/dphi-official/Datasets/master/eng_grad_emp_salary/training_set_label.csv" )
```

```
ed.head()
```

```
ed.Degree.value_counts()
```

```
ed=ed[['Gender','10percentage','12graduation','12percentage','CollegeTier','Degree','collegeGPA','CollegeCityTier','GraduationYear','English',	'Logical',	'Quant',	'Domain','ComputerProgramming',	'ElectronicsAndSemicon',	'ComputerScience',	'MechanicalEngg',	'ElectricalEngg',	'TelecomEngg',	'CivilEngg',	'conscientiousness',	'agreeableness',	'extraversion',	'nueroticism',	'openess_to_experience',	'Salary']]
```

```
ed['Gender']=ed['Gender'].apply(lambda x: 1 if x=='m' else 0)
ed=pd.get_dummies(ed,columns=['Degree'])
ed.head()
```

```
ed.isnull().sum()
```

```
ed.info()
```

```
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
```

```
x=ed.drop('Salary',axis=1)
y=ed.Salary
```

```
x_train, x_test, y_train, y_test=train_test_split(x,y,test_size=0.1,random_state=562)
```

```
Model=LinearRegression()
Model.fit(x_train,y_train)
```

```
pred=Model.predict(x_test)
```

```
from sklearn.metrics import mean_squared_error
np.sqrt(mean_squared_error(y_test,pred))
```
